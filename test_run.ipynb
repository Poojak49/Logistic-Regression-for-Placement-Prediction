```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/placement.csv')
```

```python
df
```

```python
df.head()
```

```python
df.info()
```

```python
df = df.iloc[:,1:]
```

```python
df
```

```python
import matplotlib.pyplot as plt
```

My Placement status colum is catagorical i want to convert that into numerical


```python
df["Placement Status"] = df["Placement Status"].replace({'Yes': 1, 'No': 0}).astype(int)
```

```python
df
```

```python
plt.scatter(df['CGPA'],df['IQ (Score)'],c=df['Placement Status'])
```

```python
x = df.iloc[:,0:2]
y = df.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split

train_test_split(x,y,test_size=0.1)
```

Now we got, x of training and test set.
y of training and test set
To store these 4 set into variables use following code.

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.1)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

Now scale the values

```python
from sklearn.preprocessing import StandardScaler
```

Make object

```python
Scaler = StandardScaler()
```

Now Scale

```python
x_train = Scaler.fit_transform(x_train)
```

```python
x_train
```

```python
x_test = Scaler.transform(x_test)
```

```python
x_test
```

Now Train the model
Logistic Regression will use

```python
from sklearn.linear_model import LogisticRegression
```

Make Object

```python
clf = LogisticRegression()
```

Fit use to train model

Model training step




```python
clf.fit(x_train,y_train)
```

Model Evaluation

Helpto get accuracy

```python
y_pred = clf.predict(x_test)
```

```python
y_pred
```

Now check the differences between prediction and your test set.

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

To see pattern which our algorithm used for prediction use following code

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(x_train, y_train.values, clf=clf, legend=2)
```

Everything is in object so pickle covert that into file. We can use that file in any env.
